### Data Preprocessing

```python
df1['Attrition']
```

```python
def attrition(x):
    if x in 'Yes':
        return 1
    else:
        x == 'No'
        return 0
```

```python
print((attrition('Yes'), attrition('No')))

df1['Attrition'] = df1['Attrition'].apply(attrition)

df1.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
df3 = pd.get_dummies(df1,dtype='int64')
df3.head()
```

#### Check the shape of our new dataset

```python
df3.shape
```

#### Print unique values in our new dataframe

```python
for column in df3.columns:
    unique_values = df3[column].unique()
    print(f"Number of unique elements in {column} are: {unique_values}")
```

#### Split the columns into input and target variables (X and y)

```python
X =  df3.drop(columns = ['Attrition'])
y = df3['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns = X.columns
```

```python
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns = xcolumns)
```

```python
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 100)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print('Accuracy score of our model is :',model.score(X_train,y_train))
y_pred = model.predict(X_test)
print('Confusion matrix is :\n',metrics.confusion_matrix(y_test,y_pred))
print('Classification report is:\n',metrics.classification_report(y_test,y_pred))
print('Cross Validation Score is:',cross_val_score(LogisticRegression(), X_test, y_test, cv=2))
```
